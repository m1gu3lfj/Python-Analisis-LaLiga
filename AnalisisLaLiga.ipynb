## Importar librerías de Python

```
import numpy as np
import pandas as pd
```

## Lectura de los datasets

```
df19 = pd.read_csv("season-1819_csv.csv")
df18 = pd.read_csv("season-1718_csv.csv")
df17 = pd.read_csv("season-1617_csv.csv")
df16 = pd.read_csv("season-1516_csv.csv")
df15 = pd.read_csv("season-1415_csv.csv")
df14 = pd.read_csv("season-1314_csv.csv")
df13 = pd.read_csv("season-1213_csv.csv")
df12 = pd.read_csv("season-1112_csv.csv")
df11 = pd.read_csv("season-1011_csv.csv")
df10 = pd.read_csv("season-0910_csv.csv")
```

```
print("Datos temporada 18-19:", df19.shape)
print("Datos temporada 17-18:", df18.shape)
print("Datos temporada 16-17:", df17.shape)
print("Datos temporada 15-16:", df16.shape)
print("Datos temporada 14-15:", df15.shape)
print("Datos temporada 13-14:", df14.shape)
print("Datos temporada 12-13:", df13.shape)
print("Datos temporada 11-12:", df12.shape)
print("Datos temporada 10-11:", df11.shape)
print("Datos temporada 09-10:", df10.shape)
```

## Selección de variables

```
df19 = df19.iloc[:,2:22]
df18 = df18.iloc[:,2:22]
df17 = df17.iloc[:,2:22]
df16 = df16.iloc[:,2:22]
df15 = df15.iloc[:,2:22]
df14 = df14.iloc[:,2:22]
df13 = df13.iloc[:,2:22]
df12 = df12.iloc[:,2:22]
df11 = df11.iloc[:,2:22]
df10 = df10.iloc[:,2:22]
dft = pd.concat([df19, df18, df17, df16, df15, df14, df13, df12, df11, df10])
dft.info()
```

## Transformación de datos

```
dft.FTR[dft.FTR == 'H'] = 0
dft.FTR[dft.FTR == 'D'] = 1
dft.FTR[dft.FTR == 'A'] = 2
dft.HTR[dft.HTR == 'H'] = 0
dft.HTR[dft.HTR == 'D'] = 1
dft.HTR[dft.HTR == 'A'] = 2
dft.FTR = dft.FTR.astype('int64')
dft.HTR = dft.HTR.astype('int64')
```

```
dft.info()
```

## Estadísticas generales Local-Visitante

```
total_partidos = dft.shape[0]

ganados_local = dft[dft.FTR == 0]['FTR'].shape[0]
empate = dft[dft.FTR == 1]['FTR'].shape[0]
ganados_visitante = dft[dft.FTR == 2]['FTR'].shape[0]

goles_local = dft.FTHG.sum()
goles_visitante = dft.FTAG.sum()
remates_local = dft.HS.sum()
remates_visitante = dft.AS.sum()
rempor_local = dft.HST.sum()
rempor_visitante = dft.AST.sum()

faltas_local = dft.HF.sum()
faltas_visitante = dft.AF.sum()
tarama_local = dft.HY.sum()
tarama_visitante = dft.AY.sum()
tarroj_local = dft.HR.sum()
tarroj_visitante = dft.AR.sum()

#Ganados, empatados y perdidos
mensaje = "Partidos ganados por el equipo local:      {:5} ({:5.2f}%)"
print(mensaje.format(ganados_local, round((ganados_local/total_partidos) * 100, 2)))
mensaje = "Partidos empatados:                        {:5} ({:5.2f}%)"
print(mensaje.format(empate, round((empate/total_partidos) * 100, 2) ))
mensaje = "Partidos ganados por el equipo visitante:  {:5} ({:5.2f}%)"
print(mensaje.format(ganados_visitante, round((ganados_visitante/total_partidos) * 100,2)))

#Goles y remates
mensaje = "Goles marcados por el equipo local:        {:5} ({:5.2f} por partido)"
print(mensaje.format(goles_local, round(goles_local/total_partidos, 2)))
mensaje = "Goles marcados por el equipo vistante:     {:5} ({:5.2f} por partido)"
print(mensaje.format(goles_visitante, round(goles_visitante/total_partidos, 2)))
mensaje = "Remates equipo local:                      {:5} ({:5.2f} por partido)"
print(mensaje.format(remates_local, round(remates_local/total_partidos, 2)))
mensaje = "Remates equipo vistante:                   {:5} ({:5.2f} por partido)"
print(mensaje.format(remates_visitante, round(remates_visitante/total_partidos, 2)))
mensaje = "Remates a puerta equipo local:             {:5} ({:5.2f} por partido)"
print(mensaje.format(rempor_local, round(rempor_local/total_partidos, 2)))
mensaje = "Remates a puerta equipo vistante:          {:5} ({:5.2f} por partido)"
print(mensaje.format(rempor_visitante, round(rempor_visitante/total_partidos, 2)))

#Faltas
mensaje = "Faltas equipo local:                       {:5} ({:5.2f} por partido)"
print(mensaje.format(faltas_local, round(faltas_local/total_partidos, 2)))
mensaje = "Faltas equipo vistante:                    {:5} ({:5.2f} por partido)"
print(mensaje.format(faltas_visitante, round(faltas_visitante/total_partidos, 2)))
mensaje = "Tarjetas amarillas equipo local:           {:5} ({:5.2f} por partido)"
print(mensaje.format(tarama_local, round(tarama_local/total_partidos, 2)))
mensaje = "Tarjetas amarillas equipo vistante:        {:5} ({:5.2f} por partido)"
print(mensaje.format(tarama_visitante, round(tarama_visitante/total_partidos, 2)))
mensaje = "Tarjetas rojas equipo local:               {:5} ({:5.2f} por partido)"
print(mensaje.format(tarroj_local, round(tarroj_local/total_partidos, 2)))
mensaje = "Tarjetas rojas equipo vistante:            {:5} ({:5.2f} por partido)"
print(mensaje.format(tarroj_visitante, round(tarroj_visitante/total_partidos, 2)))
```
